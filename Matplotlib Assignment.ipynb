**Q1: What is Matplotlib? Why is it used? Name five plots that can be plotted using the Pyplot module of
Matplotlib.**

A1.

Matplotlib is a comprehensive plotting library for the Python programming language and its numerical mathematics extension, NumPy. It provides an object-oriented API for embedding plots into applications or creating interactive visualizations.

- Data Visualization: It helps in visualizing data in a variety of formats, making it easier to understand and analyze.
- Customizable Plots: Users can customize the appearance of plots, including colors, labels, and markers.
- Wide Range of Plot Types: It supports a variety of plot types and styles for different data visualization needs.
- Integration: Works well with other libraries like Pandas and NumPy for data manipulation and analysis.
- Interactivity: Provides tools for creating interactive plots and dashboards.

Five Plots That Can Be Plotted Using the Pyplot Module of Matplotlib:

- Line Plot: Displays data points connected by lines. Useful for showing trends over time.
- Bar Plot: Represents data with rectangular bars. Suitable for comparing quantities among different groups.
- Histogram: Shows the distribution of data by grouping values into bins. Useful for understanding the frequency distribution.
- Scatter Plot: Displays individual data points. Useful for visualizing the relationship between two variables.
- Pie Chart: Shows proportions of a whole as slices of a pie. Useful for showing percentage breakdowns.

**Q2: What is a scatter plot? Use the following code to generate data for x and y. Using this generated data
plot a scatter plot.**
Note: Also add title, xlabel, and ylabel to the plot.

```python
import numpy as np
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
```

A2.

A scatter plot is a type of data visualization that displays individual data points on a two-dimensional plane. It is used to examine the relationship or correlation between two continuous variables. Each point on the scatter plot represents a pair of values, one for the x-axis and one for the y-axis.

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize= (10,6))
plt.scatter(x= x, y= y)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Scatter Plot')
plt.show()
```

**Q3: Why is the subplot() function used? Draw four line plots using the subplot() function.
Use the following data:**

```python
import numpy as np

x1 = np.array([0, 1, 2, 3, 4, 5]) 
y1 = np.array([0, 100, 200, 300, 400, 500])

x2 = np.array([0, 1, 2, 3, 4, 5])
y2 = np.array([50, 20, 40, 20, 60, 70])

x3 = np.array([0, 1, 2, 3, 4, 5]) 
y3 = np.array([10, 20, 30, 40, 50, 60])

x4 = np.array([0, 1, 2, 3, 4, 5]) 
y4 = np.array([200, 350, 250, 550, 450, 150])
```

A3.

The subplot() function in Matplotlib is used to create multiple subplots within a single figure. It helps in organizing different plots in a grid layout, making it easier to compare multiple plots side by side or in a specific arrangement.

```python
data = [(x1, y1), (x2, y2), (x3, y3), (x4, y4)]

plt.figure(figsize= (10, 12))

for i in range(len(data)):
    plt.subplot(4, 1, i+1)
    plt.plot(data[i][0], data[i][1])
    plt.title(f'Line plot for line {i+1}')

plt.tight_layout()
plt.show()
```

**Q5: What is a box plot? Why is it used? Using the following data plot a box plot.**

```python
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)
```

A5.

A box plot (or box-and-whisker plot) is a graphical representation of data that shows the distribution of the data through their quartiles. It provides a summary of the data's central tendency, variability, and presence of outliers.

```python
plt.figure(figsize=(14,6))
plt.boxplot([box1, box2], vert= False)
plt.xlabel('Values')
plt.ylabel('Groups')
plt.title('Horizontal Box Plot of Two Groups')
plt.show()
```
